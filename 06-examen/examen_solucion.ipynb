En este caso habría que entender el valor económico de tener en claro que monto se va a cobrar para saber si justifica o no el modelo. En caso de la preparación de la data en este caso habría que repetir eliminar los nulos y outliers. El modelo debería predecir una valor continuo, el monto a cobrar mensual. Para evaluarlo deberíamos hacer un back testing evaluando meses que conocemos sus resultados vs el valor del modelo. El deployment podría ser 1 vez al mes, no necesita ser real time.
